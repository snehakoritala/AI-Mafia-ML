```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.linear_model import LinearRegression
```

```python
x=pd.read_csv('Downloads/Linear_X_Train.csv')
y=pd.read_csv('Downloads/Linear_Y_Train.csv')
```

```python
plt.scatter(x,y,color='red',marker='+')
plt.show()
```

```python
m1=LinearRegression()
m1.fit(x,y)
```

```python
test=pd.read_csv('Downloads/Linear_X_Test.csv')
output=m1.predict(test)
coeff=m1.coef_
intercept=m1.intercept_
print('Coefficient:',coeff)
print('Intercept:',intercept)
print('Model score:',m1.score(x,y))
```

```python
plt.scatter(x,y,marker='+',color='red',label='Actual')
plt.plot(test,output,color='blue',label='Predicted')
plt.legend()
plt.show()
```
